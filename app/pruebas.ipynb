```python
import pandas as pd
import numpy as np
Number =list(range(1,11))
Name = [ 'raul','andre','patricia','flora','mateo','braulio','jhonny','andrea','jorge','luis']
City = ['la paz','el alto','el alto','potosi','santa cruz','cocha','beni','pando','tarija','sucre']
columns=['Number','Name','City']
dataset = pd.DataFrame({'Number':Number,'Name':Name,'City':City})
Gender = pd.DataFrame({'Gender':['M','F','M','M','F','F','F','M','M','F']})
Height = pd.DataFrame(np.random.randint(12,175,size=(12,1)))
Weight = pd.DataFrame(np.random.randint(50,110,size=(12,1)))
dataset['Gender']=Gender
dataset['Height']=Height
dataset['Weight']=Weight

dataset.set_index('Number')
dataset
df = dataset.copy(deep=True)
df
```

```python
df.describe()
```

```python
df.corr(numeric_only=True)
```

```python
df.count()
```

```python
# Manejo de grupos
df.groupby('City')['Gender'].count()
```

```python
print(dataset.groupby('City')['Gender'].groups)
```

```python
print(dataset.groupby(['City','Gender']).groups)
```

```python
# Iterando en los grupos
grouped = df.groupby('Gender')
for name,group in grouped:
    print(name)
    print(group)
    print("\n")
```

```python
# Seleccionar solo 1 grupo
grouped = df.groupby('Gender')
print(grouped.get_group('M'))
```

```python
## Agregaciones
grouped = df.groupby('Gender')
print(grouped['Height'].mean())
print(grouped['Weight'].mean())
print(grouped.size())
print(grouped['Weight'].agg([np.sum,np.mean,np.std]))
```

```python
# Transformaciones
df = df.set_index(['Number'])
df
```

```python
grouped = df.groupby('Gender')
score = lambda x: (x-x.mean())/x.std()*10 if isinstance(x, (int, float)) else x
grouped.transform(score)
```

```python
# Filtros
print(df.groupby('City').filter(lambda x: len(x)>=2))
```
