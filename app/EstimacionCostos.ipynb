# Costos de mantenimiento

Los costos detallados en el gráfico 1 se originan en el Estudio Financiero y la posterior Aclaración de Valores de Reposición de Activos Electromecánicos, desarrollados por el fabricante Doppelmayr Seilbahnen GmbH, basándose en la estimación de la vida útil de los elementos del sistema electromecánico diferenciando las piezas o elementos que requieren o no sustitución, lo que permitió:
•	Establecer un intervalo de reposición del 100% a piezas o elementos que se sustituyen.

•	Y en aquellas piezas que tengan un coste de mantenimiento dar un porcentaje del valor original equivalente al coste del mismo.

Por lo tanto los costos de la figura 1, engloban los costos de los servicios de mantenimiento y los costos de los bienes requeridos (como ser: repuestos, equipos, lubricantes, combustibles, insumos y material de limpieza) para realizar los mantenimientos preventivos y predictivos.

Con el propósito de reducir el impacto del costo que suponen los picos expuestos en la curva del gráfico 1, se incrementó el alcance del mantenimiento basado en la condición a los activos permitiendo extender los periodos de los mantenimientos mayores del STC, generando ahorros significativos en costos. Basado en el monitoreo continuo de los equipos críticos utilizando tecnología avanzada, se ha logrado optimizar los mantenimientos preventivos mayores ejecutándolos solo cuando es necesario, maximizando la vida útil de los activos y reduciendo los tiempos de inactividad.

La curva de costos de la presenta una reducción en sus picos en las diferentes gestiones, reduciendo considerablemente los costos de mantenimiento.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.interpolate import make_interp_spline


# Crear una lista de años desde 2022 hasta 2050
years = list(range(2022, 2050))

# Crear una lista de costos correspondientes a los puntos en el gráfico
costs = [65, 108, 34, 31,42,48, 75, 48, 71, 58, 49, 68, 47, 79, 58, 79, 65, 75, 41, 59,111, 120, 59, 68, 51, 95, 41, 68]

# Crear el DataFrame
data = {'Year': years, 'Cost': costs}
df = pd.DataFrame(data)

# Crear el modelo de regresión lineal
X = np.array(df['Year']).reshape(-1, 1)
y = np.array(df['Cost'])

model = LinearRegression()
model.fit(X, y)

# Obtener la ecuación de la regresión lineal
intercept = model.intercept_
slope = model.coef_[0]
print(f"Ecuación de la regresión lineal: y = {slope:.2f}x + {intercept:.2f}")

# Hacer predicciones para los próximos 20 años (2051-2070)
future_years = np.array(range(2051, 2071)).reshape(-1, 1)
future_costs = model.predict(future_years)

# Crear el DataFrame para los años futuros
future_data = {'Year': range(2051, 2071), 'Cost': future_costs}
df_future = pd.DataFrame(future_data)

# Combinar los datos actuales con las predicciones futuras
df_combined = pd.concat([df, df_future], ignore_index=True)

# Suavizar las líneas utilizando splines
x_smooth = np.linspace(df_combined['Year'].min(), df_combined['Year'].max(), 300)
spl = make_interp_spline(df_combined['Year'], df_combined['Cost'], k=3)
y_smooth = spl(x_smooth)

# Crear la gráfica
plt.figure(figsize=(12, 6))
plt.plot(x_smooth, y_smooth, color='black', label='Tendencia suavizada')
plt.scatter(df['Year'], df['Cost'], color='black', label='Datos actuales')
plt.scatter(df_future['Year'], df_future['Cost'], color='red', marker='x', label='Proyecciones futuras')
plt.title('Costos por año con Proyecciones Futuras')
plt.xlabel('Año')
plt.ylabel('Costo')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)

# Añadir los valores de los puntos en el gráfico
for i in range(len(df)):
    plt.text(df['Year'][i], df['Cost'][i], df['Cost'][i], fontsize=9, ha='center', va='bottom')
for i in range(len(df_future)):
    plt.text(df_future['Year'][i], df_future['Cost'][i], f"{df_future['Cost'][i]:.0f}", fontsize=9, ha='center', va='bottom')

# Mostrar la gráfica
plt.show()
```
